```python
import numpy as np
import pandas as pd

table = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8", match="1950—2019")
```

```python
table[0].head()
```

Визначте кількість рядків та стовпців у датафреймі (атрибут shape)

```python
print(np.shape(table))
```

Замініть у таблиці значення "—" на значення NaN

```python
table = table[0].replace(to_replace='—', value=np.nan, regex=True)
table
```

Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
table.dtypes
```

Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
table_with_null = table.fillna(0)
table_with_null
```

Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
print(table.isnull().sum()/len(table)*100)
print(table_with_null.isnull().sum()/len(table_with_null)*100)
```

Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
without_country = table.drop(table.index[-1])
without_country
```

Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
mean_table = table.fillna(table.mean(numeric_only=True))
mean_table
```

Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
table_with_null["2019"] = pd.to_numeric(table_with_null["2019"])
table_with_null[table_with_null["2019"]>table_with_null["2019"].mean()]["Регіон"]
```

У якому регіоні була найвища народжуваність у 2014 році?

```python
table_with_null["2014"] = pd.to_numeric(table_with_null["2014"])
table_with_null[table_with_null["2014"] == table_with_null["2014"].max()]["Регіон"]
```

Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
import matplotlib.pyplot as plt

plt.bar(table_with_null["Регіон"], table_with_null["2019"])

plt.xlabel("Регіон")
plt.ylabel("Народжуваність у 2019 році")
plt.title("Народжуваність по регіонах у 2019 році")

plt.xticks(rotation=90)

plt.show()
```

```python
years = ['1950', '1960', '1970', '1990', '2000', '2012', '2014', '2019']

for index, row in table.iterrows():
    region = row['Регіон']
    plt.plot(years, row[years], label=region)

plt.xlabel('Роки')
plt.ylabel('Населення')
plt.title('Зміна населення в регіонах з часом')
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns

data_long = pd.melt(table, id_vars=['Регіон'], var_name='Рік', value_name='Показник')

data_long = data_long.dropna()

sns.scatterplot(data=data_long, x='Рік', y='Показник', hue='Регіон', palette='Set2')
plt.grid(True)
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')

plt.show()
```
